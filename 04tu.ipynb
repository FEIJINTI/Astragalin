```python
import numpy as np
import matplotlib.pyplot as plt
from utils import read_envi_ascii
```

```python
data = read_envi_ascii('data/biaotu.txt')
huangqi = data['huangqi']
zazhi1 = data['zazhi1']
zazhi2 = data['zazhi2']
beijing = data['beijing']
gancaopian = data['gancaopian']
hongqi = data['hongqi']
```
