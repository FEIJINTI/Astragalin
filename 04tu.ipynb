```python
import numpy as np
import matplotlib.pyplot as plt
from utils import read_envi_ascii
```

```python
Astragalin_data_path = 'data/huangqizazhi.txt'
data = read_envi_ascii(Astragalin_data_path)
hongqi_data_path = 'data/hongqi.txt'
hongqi_data = read_envi_ascii(hongqi_data_path)
astragalin = data['hongqi'][50:100,:]
gancaopian = hongqi_data['gancaopian'][50:100,:]
hongqi = hongqi_data['hongqi'][50:100,:]
```

```python
bands_file_path = 'data/Camera2_ref.txt'
with open(bands_file_path, 'rb') as f:
    bands = []
    for k in f:
        a = float(k[:-3])
        bands.append(a)
    bands = np.array(bands)
```

```python
# astragalin
astragalin = astragalin.T
hongqi = hongqi.T
gancaopian = gancaopian.T
fig,ax = plt.subplots()
ax.plot(astragalin,label='astragalin')
ax.plot(gancaopian,label='gancaopian')
ax.plot(hongqi,label='hongqi')
plt.xlabel('Wavelength(nm)')
plt.ylabel("intensity")
plt.legend()
plt.savefig('data/fig1.png')
```
